```python
# Импортируем нужные нам библиотеки
import pandas as pd
```

```python
# Открываем наш файл csv
df = pd.read_csv('/content/Список женской одежды с Садовода.csv')
```

```python
# Проверим в каких столбцах есть пустые значения
df.isnull().sum()
```

```python
# Удалим все пустые значения
df1 = df.dropna()
```

```python
# Снова проверяем пустые значения после очистки
df1.isnull().sum()
```

```python
#  Сохраним новый чистый датасет
df1.to_csv('Чистый список женской одежды.csv')
```
